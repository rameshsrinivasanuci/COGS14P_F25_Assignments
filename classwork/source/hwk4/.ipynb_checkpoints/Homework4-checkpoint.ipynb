1. Write a function called **bank** computes the **balance** of an account with annually compounded **interest** until the balance reaches a **target** balance.  It also keeps track of the number of years (i.e., the number of repetitions of the while loop) to reach the target.  

- The input arguments should be the initial value of **balance** and **interest** and the **target** at which to stop calculating. Keep your arguments in that order.
- The function should return the number of **years** and a list containing the **balance** every year starting from the year 0 starting balance.
- The first variable returned should be the number of **years** and the second variable returned should be the yearly **balance**    

Your function should work with **interest** provided as a percentage because that is how humans think.  Convert to a fraction by dividing by 100. 

Hint: When returning two variables from a function the return should list the variables separated by a comma
e.g.,
`return a,b`

```python
### BEGIN SOLUTION
def bank(balance,interest,target):
    interest = interest/100 
    balance_yearly = [balance]
    nyears = 0 
    while balance <= target: #Execute the block as long as my balance is less than or equal to target!  
        balance = balance+interest*balance
        nyears = nyears+1
        balance_yearly.append(balance)
    return nyears,balance_yearly
### END SOLUTION
```

```python
interest = 6
balance = 5000
target = 100000
nyears,yearly_balance = bank(balance,interest,target)
assert nyears == 52
assert yearly_balance[0] == balance
assert len(yearly_balance) == 53
```

2. COLLATZ CONJECTURE

- This is one of these weird things with numbers that remains unproven, but always works. Collatz had this insight in 1939, but could not prove it, hence it is a conjecture. 

- Start with any positive integer n.    
- Follow this rule: 
1. if the integer is even, the next integer is one half of the current integer. 
2. if the integer is odd, the next integer is 3 times the integer plus one. 
3. Eventually the sequence of integers will reach 1. This always works! 

2a. Write a function that takes as an input argment the positive integer n and returns the sequence of numbers that terminates at 1 as a list or array.
- A reminder that the % operator returns the remainder.  This can be helpful in testing for even or odd numbers. 


```python
 ### BEGIN SOLUTION
def collatz(n): 
    sequence = [n] 
    while n != 1:
        if n%2 == 0:
            n = n/2
        else:
            n = 3*n+1
        sequence.append(n)
    return sequence

### END SOLUTION
```

```python
n = 7
col_seq = collatz(n)
assert len(col_seq) == 17
assert col_seq[0] == 7
assert np.max(col_seq) == 52
```

3. Create a function **convertgrade**  which takes input arguments a numeric grade, and a gradingscale given as a list of string letter grades, and a corresponding list of numeric lower cut off. 
For example, if 
    
        numeric = 92
        grades = ['A','B','C','D','F']
        cutoff = [90,80,70,60,0]

The function should return the letter grade corresponding to the numeric grade. For this example, the function should return a variable with the letter grade 'A'

**The function should be able to work with any arbitrary scale without making any changes to the function.**

So the same function **without any changes** should process these data

        numeric = 92
        grades = ['A','A-','B+','B','B-','C+','C','C-','F']
        cutoff = [94,90,86,82,78,74,70,65,0]

and return a letter grade of 'A-'

The function should work with any *arbitrary* grade system with corresponding numeric lower cutoffs.  

Hint: `len(mylist)` returns the length of mylist
Hint: My test cases assume that the input arguments for your function are in the order **numeric**, **grades**, and **cutoff**.  The length of grades and cutoff can be assumed to be the same. 


```python

### BEGIN SOLUTION
def convertgrade(numeric,grades,cutoff):
    ngrades = len(grades)
    for j in range(ngrades-1,-1,-1):
        if numeric >= cutoff[j]:
            letter = grades[j]
    return letter
### END SOLUTION
```

```python
numeric = 92
grades = ['A','B','C','D','F']
cutoff = [90,80,70,60,0]
letter = convertgrade(numeric,grades,cutoff)
assert letter == 'A'
```

```python
numeric = 62
grades = ['Q','R','S','T','U','V']
cutoff = [90,80,70,60,50,0]
letter = convertgrade(numeric,grades,cutoff)
assert letter == 'T'
### BEGIN HIDDEN TESTS 
numeric = 92
grades = ['A','A-','B+','B','B-','C+','C','C-','F']
cutoff = [94,90,86,82,78,74,70,65,0]
letter = convertgrade(numeric,grades,cutoff)
assert letter == 'A-'
### END HIDDEN TESTS
```

4.  Write an interactive game, where the task is to guess a 2 digit number within 3 guesses.

- The game should draw a random number between 10 and 99 (inclusive), which is the answer. Use default_rng to create a random number generator and use the `integers` method to draw your random number.   

- The program asks the player to input a number, and print out "too low" if the guess is less than the correct answer and "too high" if the guess is greater than the correct answer.  

- The game ends if the player guesses the correct answer.  In this case, the game should print out "winner:" and the correct answer.  

- If the player guesses 3 times incorrectly, the game should exit with the message "Too Many Guesses". 

Hint: The `input` statement returns a variable as type **string**.  You can use the function `int` to convert to an integer.  

Hint: If you set up the random number seed as 11 when you create the random number generate using `default_rng`, the correct guess is 22.  This can help for testing.  

```python
### BEGIN SOLUTION
from numpy import random
rng = random.default_rng(seed = 11)
answer = rng.integers(10,100,1)
nguess = 0 
error = 1
while (nguess < 3) &  (error != 0): 
    guess = input('Guess the number:')
    guess = int(guess)
    nguess = nguess + 1
    error = guess - answer
    if error > 0:
        print("too high")
    elif error < 0:
        print("too low")
    else:
        print("winner: ", answer)
if (nguess >= 3) & (error != 0):
    print('Too Many Guesses')

### END SOLUTION

```
